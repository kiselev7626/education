```python
direction = str(input('choose direction - encrypt (e) or decrypt (d)'))
language = str(input('choose language - Russian (r) or / English (e)'))
pre_step = int(input('choose step')) # to the right
text_in = str(input('enter text'))

alphabet = ''
if language == 'r':
    alphabet += 'АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
if language == 'e':
    alphabet += 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

if direction == 'd':
    pre_step *= -1
```

```python
#alphabet = 'АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ'
#text_in = 'Умом Россию не понять'
#text_in = 'Фнпн Спттйя ож рпоауэ'
text_out = ''

pre_step = 0
step = pre_step % len(alphabet) # ниввелирую число больше чем длина алфавита

def change_alpha(text_in_1, i_1, alphabet_1):
    alphabet_1.find(text_in_1[i_1].upper()) # номер символа в алфавите
    if alphabet_1[alphabet_1.find(text_in_1[i_1].upper())] == text_in_1[i_1]: # если символ большой равен символу маленькуому, значит буква большая
        new_symbol = alphabet_1[(alphabet_1.find(text_in_1[i_1].upper()) + step) % len(alphabet_1)]
    else:
        new_symbol = alphabet_1[(alphabet_1.find(text_in_1[i_1].upper()) + step) % len(alphabet_1)].lower()
    #print(new_symbol, end='')
    return new_symbol

for i in range(len(text_in)):
    if text_in[i].isalpha() == True:
        new_symbol = change_alpha(text_in, i, alphabet)
    else:
        new_symbol = text_in[i]
    text_out += new_symbol

print(text_out)
```
